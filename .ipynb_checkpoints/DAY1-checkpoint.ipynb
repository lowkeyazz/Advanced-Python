```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
T = pd.read_csv("plot.csv")
T['3.000000'] = T['3.000000'].replace({'1.#INF00': 900})
T = T['3.000000'].tolist()
T = np.insert(T,0,3.0)
```

```python
def split_list_by_index(T):
    cpu = T[0::2] 
    band = T[1::2] 
    return cpu, band
```

```python
T = np.array([float(value.replace("'", "")) for value in T])
T
```

```python
cpu , band = split_list_by_index(T)
```

```python
def split_list_into_halves(lst):
    m = len(lst) // 2
    hlf = lst[:m]
    shlf = lst[m:]

    matrix = np.array([hlf, shlf])

    return matrix
cpu_m = split_list_into_halves(cpu)
band_m = split_list_into_halves(band)
```

```python

for row in cpu_m:
    plt.plot(row)
    

plt.xlabel('Strides')
plt.ylabel('CPU time')
plt.title('The penalty of stride')

# Show the plot
plt.show()
```

```python

for row in band_m:
    plt.plot(row)

plt.xlabel('Strides')
plt.ylabel('Bandwidth')
plt.title('The penalty of stride')

# Show the plot
plt.show()
```
